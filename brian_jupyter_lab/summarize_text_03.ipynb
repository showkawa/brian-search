```python
from langchain.document_loaders import UnstructuredPDFLoader

def load_pdf(pdf_path):
    loader = UnstructuredPDFLoader(pdf_path)
    data = loader.load()
    return data
```

```python
docs = load_pdf("./../../../doc/teams API/699402 office365devguide-陈希章-594.pdf")
```

```python
print(f"docment number {len(docs)}")
print(f"total characters {len(docs[0].page_content)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000,chunk_overlap=10)
split_docs=text_splitter.split_documents(docs)
```

```python
print(f"split documents number {len(split_docs)}")
```

```python
import os
import getpass

os.environ['OPENAI_API_KEY'] = getpass.getpass('OpenAI API Key:')
```

```python
from langchain.llms import OpenAI
from langchain.chains.summarize import load_summarize_chain
```

```python
llm=OpenAI(temperature=0)
chain = load_summarize_chain(llm, chain_type="map_reduce",verbose=True)
```

```python
chain.run(input_documents=split_docs)
```
